```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.mixture import GaussianMixture
import plotly.express as px


#беремо файл за 2019 рік
data = pd.read_csv('2019.csv')
data = data.rename(columns={'Score': 'Рівень щастя', 'GDP per capita': 'ВВП на особу', 'Social support': 'Соціальна підтримка', 'Healthy life expectancy': 'Тривалість життя', 'Freedom to make life choices': 'Свобода вибору', 'Generosity': 'Щедрість', 'Perceptions of corruption': 'Сприйняття корупції', 'Overall rank': 'Рейтинг'});
data_copy = data.copy()

print('Перші рядки:\n')
print(data.head())
print('--------------------------\n')

#загальна інформація, щоб зрозуміти структуру даних
print('Інформація по dataset:\n')
print(data.info())
print('--------------------------\n')

#середні значення та діапазони
print('Опис статистики:\n')
print(data.describe())
print('--------------------------\n')

#відбираємо числові ознаки
numbers = data.select_dtypes(include=['float64', 'int64'])
numbers2 = data_copy.select_dtypes(include=['float64', 'int64'])

#розподіл значень
numbers.hist(figsize=(12,8), color='#6c8ebf', edgecolor='black');
plt.suptitle('Розподіл ознак');
plt.show()
numbers2.hist(figsize=(10,6), color='#9dbad5', edgecolor='black');
plt.show()


#дивлюсь які фактори можуть впливати на щастя
cols = ['Рівень щастя', 'ВВП на особу', 'Соціальна підтримка', 'Тривалість життя', 'Свобода вибору']
cols_try = numbers.columns[:5]
corr = data[cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='crest');
plt.title('Кореляційна матриця');
plt.show()


# карта щастя по країнах
fig = px.choropleth(data,
                    locations='Country or region',
                    color='Рівень щастя',
                    locationmode='country names',
                    color_continuous_scale='Teal')

fig.update_layout(title='Рівень щастя по країнах (2019)');
fig.show()


#ф-ція для масштабування
def data_scale(data, type='minmax'):
    if type == 'minmax':
        scaler = MinMaxScaler()
    if type == 'std':
        scaler = StandardScaler()
    if type == 'norm':
        scaler = Normalizer()

    scaler.fit(data)
    res = scaler.transform(data)
    return res


#використ методів
scaled_minmax = data_scale(numbers, 'minmax')
scaled_std = data_scale(numbers, 'std')
scaled_norm = data_scale(numbers, 'norm')
scaled_std2 = data_scale(numbers2, 'std')


df_minmax = pd.DataFrame(scaled_minmax, columns=numbers.columns)
df_std = pd.DataFrame(scaled_std, columns=numbers.columns)
df_norm = pd.DataFrame(scaled_norm, columns=numbers.columns)
df_for_cluster = df_std.copy()

print('Мінмакс масштабування:\n')
print(df_minmax.describe())
print('--------------------------\n')
print('Стандартизація:\n')
print(df_std.describe())
print('--------------------------\n')
print('Нормалізація:\n')
print(df_norm.describe())
print('--------------------------\n')


#використовую стандартизацію для кластеризації
gmm = GaussianMixture(n_components=3)

clusters = gmm.fit_predict(df_for_cluster)

data['cluster'] = clusters;

print('Країни та їх кластери:\n')
print(data[['Country or region', 'cluster']].head())
print('--------------------------\n')


#розподілилення кластерів
plt.figure(figsize=(8,6))
sns.scatterplot(x=data['ВВП на особу'],
                y=data['Рівень щастя'],
                hue=data['cluster'],
                palette='Set2')
plt.xlabel('ВВП на особу')
plt.ylabel('Рівень щастя')
plt.title('Розподіл кластерів');
plt.show()


#картa кластерів
fig2 = px.choropleth(data,
                     locations='Country or region',
                     color='cluster',
                     locationmode='country names',
                     color_continuous_scale='Pastel')

fig2.update_layout(title='Кластери країн');
fig2.show()
```
